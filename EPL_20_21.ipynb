```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('EPL_20_21.csv')
df.head()
```

# **Age Count**

```python
plt.figure(figsize= (15,5))
plt.xticks(rotation= 45)
plt.title("Age Count", size= 30)
ax= sns.countplot(data= df, x= 'Age')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.info()
```

```python
df[df['Club'] == 'Chelsea']
```

# **Clubwise Count**

```python
plt.figure(figsize= (15,5))
plt.xticks(rotation= 45)
plt.title("Players Count According to Clubs", size= 30)
ax= sns.countplot(data= df, x= 'Club')

for bars in ax.containers:
    ax.bar_label(bars)
```

# **min/match and goal/match**

```python
df['MinPerMatch'] = df['Mins']/df['Matches'].astype(int)
df['GoalsPerMatch'] = df['Goals']/df['Matches'].astype(float)
df.head()
```

```python
df.dtypes
```

# **Penalty Count**

```python
penalty_attempt = df['Penalty_Attempted'].sum()
penalty_attempt
```

```python
penalty_scored = df['Penalty_Goals'].sum()
penalty_scored
```

```python
penalty_missed = penalty_attempt - penalty_scored
penalty_missed
```

```python
data = [penalty_scored, penalty_missed]
labels = ['Penalty Scored', 'Penalty Missed']
plt.pie(data, labels = labels, autopct = '%0.1f%%')
plt.show()
```

# **Player's Nationality**

```python
plt.figure(figsize= (15,5))
plt.xticks(rotation= 45)
plt.title("Player's Nationality", size= 30)
sns.countplot(data= df, x= 'Nationality')
```

# **Club's Average Age**

```python
num_player= df.groupby('Club').size()
data= df.groupby('Club')['Age'].sum()/num_player
data.sort_values(ascending=False)
```

```python
num_player= df.groupby('Club').size()
data= df.groupby('Club')['Age'].sum()/num_player
data = data.sort_values(ascending=False)
plt.title("Average Age (clubwise)", size= 30)
ax = data.plot(kind='bar', figsize=(15, 5), color=sns.color_palette('Set2'))

for bars in ax.containers:
    ax.bar_label(bars)
```

# **Clubwise Assists**

```python
assists_by_club =pd.DataFrame( df.groupby('Club', as_index= False)['Assists'].sum().sort_values('Assists', ascending= False))
assists_by_club
```

```python
plt.figure(figsize= (15,5))
plt.xticks(rotation= 70)
plt.title("Clubwise Assists", size= 30)
ax= sns.barplot(data= assists_by_club, x= 'Club', y= 'Assists')

for bars in ax.containers:
  ax.bar_label(bars)
```

# **Goals and Assists**

```python
plt.figure(figsize= (15,5))
total_gaoals = df['Goals'].sum()
assists= df['Assists'].sum()
data= [total_gaoals - assists, assists]
labels= ['Goals without Assists', 'Goals with Assists']
plt.pie(data, labels= labels, autopct= '%0.2f%%')
plt.show()
```
